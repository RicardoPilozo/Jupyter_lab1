# ** Bokeh **

Importar Libreria

```python
import statistics as stats
```

## 1. Agregar anotaciones 
Bokeh incluye varios tipos diferentes de anotaciones que puede usar para agregar información complementaria a sus visualizaciones.

```python

a = stats.mean(x)
b = stats.mean(y)

from bokeh.plotting import figure, output_file, show

p = figure(title="Basic Title", width=300, height=300)
p.circle([a,b], [b,a])

output_file("Grafico_1.html")

show(p)
```

## 2. bokeh.plotting
La base para cualquier visualización de datos son los datos subyacentes

```python
from bokeh.plotting import figure

x_values = x
y_values = y

p = figure()
p.circle(x=x_values, y=y_values)
```

## 3. Marcadores de dispersión
Bokeh incluye una gran variedad de marcadores para crear diagramas de dispersión. 
Por ejemplo, para renderizar marcadores de dispersión circular en un gráfico

```python
from bokeh.plotting import figure, output_file, show

# output to static HTML file
output_file("line.html")

p = figure(width=900, height=700)

# add a circle renderer with a size, color, and alpha
p.circle(x, y, size=20, color="navy", alpha=0.5)

# show the results
show(p)
```

# ** Seaborn * *

Importar Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

## 1. seaborn.scatterplot
Debe usarse cuando ambas variables son numéricas, es la scatterplot()función
ibuja un diagrama de dispersión con posibilidad de varias agrupaciones semánticas.
La relación entre xy ypuede demostrarse para diferentes subconjuntos de los datos usando el hue, sizey styleparámetros. 

```python
sns.relplot(x="Age", y="Income", data=df);
```

# 2. Relplot
Si bien los puntos se representan en dos dimensiones, se puede agregar otra dimensión al gráfico coloreando los puntos de acuerdo con una tercera variable. En seaborn, esto se denomina uso de una "semántica de tono", porque el color del punto adquiere significado

```python
sns.relplot(x="Age", y="Income", hue="Car_Ownership", data=df);
```

Si la semántica de tono es numérica (específicamente, si se puede convertir en flotante), el color predeterminado cambia a una paleta secuencial

```python
sns.relplot(x="Age", y="Income", hue="CURRENT_JOB_YRS", data=df);
```
